```python
import wfdb
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = wfdb.io.rdrecord("rec_1")
signal = data.adc()[:, 0]
```

```python
t = np.linspace(0, 20, 10_000)
noise = np.sin(t * 2 * np.pi * 50)
trend = -np.sin(t * 2 * np.pi * .025)
noisy_signal = signal + noise * 70 + trend * 40
```

```python
plt.plot(t, noisy_signal)
```

```python
np.savetxt("ecg_signal.txt", noisy_signal)
```
